## Cleaning the Data

```python
#removing unwanted columns/features (basic data cleaning)

marine = marine.drop('item_id', axis = 1)
marine = marine.drop('log_index', axis = 1)
marine = marine.drop('id', axis = 1)
marine = marine.drop('index', axis = 1)
```

### Our dataset is huge to be compatible for plotting it on the map. So we would have to resort the way of removing couple of instances from the dataset. We will begin this by looking if any row has missing data, if so then we would remove that row. As we can see from above, around 55 instances are missing from the location column. So initially we would be removing the rows that has missing data (location)

```python
marine_f = marine.dropna(how='any')
marine_f.count()

#dropping the rows that have missing instances
```

```python
#dropping the last 30000 rows(since the dataset is huge)

marine_f = marine_f.drop(marine_f.index[20000:50915])
marine_f.head()
```
